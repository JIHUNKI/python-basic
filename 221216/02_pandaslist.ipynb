```python
## 파일의 목록 출력
import os
import pandas as pd
```

```python
file_list = os.listdir("../csv/2017")
```

```python
## 파일 리스트는 출력 성공
## 파일에 12개가 존재
## 12개의 파일을하나의 데이터프레임에 결합
## "../csv/2017/" + file_list 형태로 만들어준다
## 비어있는 데이터프레임 생성


## 파일 리스트 , path 지정해놓은 변수
## 반복문 사용
## 어떤 것을 반복해야 되는가? ->  파일 리스트의 값들을 반복
## concat() 사용할 데이터 프레임을 로드
## 비어있는 데이터 프레임에 로드한 데이터 프레임 결합
file_path = "../csv/2017/"
df = pd.DataFrame()
for i in range(0,len(file_list),1):
      file= pd.read_csv(file_path + file_list[i])
      df = pd.concat([df,file], axis=0, ignore_index=True)

df = pd.DataFrame()
for i in file_list:
      file = pd.read_csv(file_path + i)
      df = pd.concat([df,file], axis = 0).reset_index(drop=True)

df
```
