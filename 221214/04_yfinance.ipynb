```python
import yfinance as yf
import pandas as pd
```

```python
CPNG = yf.download("CPNG",start = "2021-01-01")
```

```python
CPNG.head()
```

```python
## 소수점 2번째 자리까지 두고 반올림처리
CPNG[["Open","High","Low","Close"]] =CPNG[["Open","High","Low","Close"]].round(2)
```

```python
CPNG.isna().sum()
```

```python
## 등락폭, 등락율 파생변수 생성
## 등락폭 = 종가 - 시가
## 등락율 = 등락폭 / 사가 *100
CPNG['등락폭'] = CPNG['Close'] - CPNG['Open']
CPNG['등락율'] = CPNG['등락폭'] / CPNG['Open'] *100
```

```python
CPNG.head()
```

```python
## 시각화
import matplotlib.pyplot as plt
```

```python
plt.plot(CPNG["등락율"])
 ## index 값이 date로 되어있어 x값에 자동으로 index값이 대입
```

```python
## 그래프 4개 출력
## subplot(행의수, 열의수, 선택영역)
## ex) subplot(2,2,1) -> 2행2열로 나누워 1번영역을 선택힌디.
plt.subplot(2,2,1)
plt.plot(CPNG["Open"].values)  ## .values 로 인덱스가 123.. 순으로나오는 스리즈로 넣는다.
plt.subplot(2,2,2)
plt.plot(CPNG["Close"].values)
plt.subplot(2,2,3)
plt.plot(CPNG["Volume"].values)
plt.subplot(2,2,4)
plt.plot(CPNG["등락율"].values)
plt.show()
```
