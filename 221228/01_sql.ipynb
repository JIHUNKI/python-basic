1. emp라는 변수의 sql에 있는 emp table의 전체데이터를 를 데이터프레임으로 저장

## pandas 복습
1. 새로운 파생변수 생성 job_code
        -salesman 은 1, manager는 2, 그 외의 직업은 3
        -컬럼은 EMPNO, ENAME, job_code만 출력

```python
emp['JOB']

## apply, def
def change(x):
    if x == "SALESMAN":
        return 1
    elif x== "MANAGER":
        return 2
    else:
        return 3

emp['JOB'].apply(change)
## apply lambda
emp['JOB'].apply(lambda x: 1 if (x=="SALESMAN") else (2 if(x=="MANAGER") else 3))

## for문
for i in emp_copy.index:
    if emp_copy.loc[i,"JOB"] =='SALESMAN':
        emp_copy.loc[i,"job_code"]=1
    elif emp_copy.loc[i,"JOB"] =="MANAGER":
        emp_copy.loc[i,"job_code"]=2
    else:
        emp_copy.loc[i,"job_code"]=3
```

```python
emp_copy[["EMPNO","ENAME","job_code"]]
```

```python
## 위의 결과를 sql 쿼리문을 이용하여 생성

sql = """ 
    select EMPNO,ENAME,JOB,
    CASE
        when JOB ='SALESMAN'
        then 1
        when JOB = 'MANAGER'
        then 2
        else 3
    END as job_code
    from EMP
    """
sql_class.excuteAll(sql)
```

### pandas
1. emp 데이터에서 파생변수 H/L 생성
2. SAL이 2000이상이면 HIGH, 2000미만이면 LOW

```python
def mutate(x):
    if x >= 2000:
        return 'HIGH'
    else :
        return'LOW'
emp_copy["H/L"] = emp_copy["SAL"].apply(mutate)
emp_copy["H/L"]
```

```python
for i in emp_copy.index:
    if emp_copy.loc[i,"SAL"] >=2000:
        emp_copy.loc[i,"H/L_2"]="HIGH"
    else:
        emp_copy.loc[i,"H/L_2"]="LOW"
emp_copy["H/L_2"]
```

```python
### sql 쿼리문
sql = """ 
    select SAL,
    if(SAL >=2000,"HIGH","LOW") as `H/L`
    from emp
"""
sql_class.excuteAll(sql)

sql_2 = """ 
    select SAL,
    case 
        when SAL >=2000 then 'HIGH'
        else 'Low'
    end as `H/L`
    from emp
    """
sql_class.excuteAll(sql_2)
```

### 이중쿼리문

### pandas
1. tran_1,tran_2,tran_d_1,tran_d_2 데이터베이스에서 select문을 이용해서 데이터프레임으로 저장
2. tran_1, tran_2 단순한 행 결합(유니온 결합)
3. tran_d_1, tran_d_2, 단순한 행 결합(유니온 결합)
4. tran, tran_d를 특정 조건에 맞게 조인 결합

```python
sql_1 = """ 
        select * from tran_1
        """
tran_1 =sql_class.excuteAll(sql_1)

sql_2 = """ 
        select * from tran_2
        """
tran_2 = sql_class.excuteAll(sql_2)

sql_3 = """ 
        select * from tran_d_1
        """
tran_d_1 = sql_class.excuteAll(sql_3)

sql_4 = """ 
        select * from tran_d_2
        """
tran_d_2 =sql_class.excuteAll(sql_4)
```

```python
tran = pd.concat([tran_1,tran_2],axis=0,ignore_index=True)
```

```python
tran_d = pd.concat([tran_d_1,tran_d_2],axis=0,ignore_index=True)
```

```python
transaction = pd.merge(tran,tran_d,how='left',on='transaction_id')
transaction
```

```python
## sql 쿼리문
sql = """
        select * from tran_1
        union
        select *from  tran_2
     
        """
sql_class.excuteAll(sql)

## 위에 방법에 아래 방법을 더하면 완성
sql_2 = """ 
    select * 
    from tran_1
    left join tran_d_1
    on tran_1.transaction_id = tran_d_1.transaction_id
"""
sql_class.excuteAll(sql_2)
```

```python
## union 후 만들어진 두 테이블을 이용한 left join
sql = """ 
    select * from
    (select * from tran_1
    union
    select * from tran_2) as tran
    left join 
    (select * from tran_d_1
    union
    select * from tran_d_2) as tran_d
    on tran.transaction_id = tran_d.transaction_id
    """
sql_class.excuteAll(sql)
```
